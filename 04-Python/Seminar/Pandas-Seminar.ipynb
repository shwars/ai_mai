# Оценка распротранения COVID-19

## Загружаем данные

Будем использовать данные о заражениях COVID-19 от [университета Джона Хопкинса](https://jhu.edu/). Данные доступны [в этом репозитории GitHub](https://github.com/CSSEGISandData/COVID-19).

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (10,3) # изменяем размер рисунков по умолчанию
```

Для начала - загружаем данные с помощью `pd.read_csv`. Можно загружать сразу из интернет, либо локально из директории `data`:

```python
base_url = "https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/" # loading from Internet
base_url = "../../../data/" # loading from disk
infected_dataset_url = base_url + "time_series_covid19_confirmed_global.csv"
recovered_dataset_url = base_url + "time_series_covid19_recovered_global.csv"
deaths_dataset_url = base_url + "time_series_covid19_deaths_global.csv"
countries_dataset_url = base_url + "UID_ISO_FIPS_LookUp_Table.csv"
```

Для начала загрузим данные о числе заражённых:

```python
infected = pd.read_csv(infected_dataset_url)
infected.head()
```

Мы видим, что в каждой строке - данные об отдельной стране или регионе, а по столбцам идут даты. Загрузим аналогичным образом данные по выздоровлениям и смертям:

```python
recovered = pd.read_csv(recovered_dataset_url)
deaths = pd.read_csv(deaths_dataset_url)
```

## Изучаем данные

Из первых строк таблицы не очень понятна роль столбца `Province/State`. Посмотрим на все уникальные значения:

```python
infected['Province/State'].value_counts()
```

Из имён понятно, что для таких стран, как Австралия или Китай, информация разбита более детально по провинциям или городам. Например, посмотрим внимательнее на Китай:

```python
infected[infected['Country/Region']=='China']
```

## Предобработка данных 

Поскольку нам не очень интересно опускаться на уровень отдельных городов, построим агрегатные значения по странам. Для этого можно использовать `groupby`:

```python
infected = infected.groupby('Country/Region').sum()
recovered = recovered.groupby('Country/Region').sum()
deaths = deaths.groupby('Country/Region').sum()

infected.head()
```

После использования `groupby` мы получили таблицы, которые индексированы по столбцу `Country/Region`. Можем обращаться к конкретной стране с помощью конструкции `.loc`:

```python
infected.loc['Russia'][2:].plot()
recovered.loc['Russia'][2:].plot()
plt.show()
```

Обратите внимание на использование `[2:]` для удаления первых двух столбцов, содержащих геолокацию страны. Чтобы эти столбцы нам не мешали, можем их удалить:

```python
infected.drop(columns=['Lat','Long'],inplace=True)
recovered.drop(columns=['Lat','Long'],inplace=True)
deaths.drop(columns=['Lat','Long'],inplace=True)
```

## Изучаем данные

Для изучения данных по конкретной стране, опишем функцию `mkframe`, которая соберёт все три набора данных о конкретной стране вместе в одну таблицу:

```python
def mkframe(country):
    df = pd.DataFrame({ 'infected' : infected.loc[country] ,
                        'recovered' : recovered.loc[country],
                        'deaths' : deaths.loc[country]})
    df.index = pd.to_datetime(df.index)
    return df

df = mkframe('Russia')
df
```

```python
df.plot()
plt.show()
```

Используем `diff`, как в лекции, чтобы построить график прироста заболеваемости по дням:

```python
df['ninfected'] = df['infected'].diff()
df['ninfected'].plot()
plt.show()
```

Можно видеть, как в данных есть значительные колебания. Посмотрим на данные за один месяц:

```python
df[(df.index.year==2020) & (df.index.month==5)]['ninfected'].plot()
plt.show()
```

Для сглаживания данных, используем **скользящее среднее**, т.е. в каждый из дней усредняем значение по окну из 7 близлежащих дней:

```python
df['ninfav'] = df['ninfected'].rolling(window=7).mean()
df['ninfav'].plot()
plt.show()
```

Чтобы сравнить течение пандемии в разных странах, имеет смысл взять в расчет население страны, и посчитать процент заболевших. Для этого загрузим данные о странах в отдельную таблицу:

```python
countries = pd.read_csv(countries_dataset_url)
countries
```

Для некоторых стран в датасете есть данные по населению как стран целиком, так и отдельных районов - поэтому для извлечения правильной информации по населению надо сделать выборку:

```python
countries[(countries['Country_Region']=='US') & countries['Province_State'].isna()]
```

Для нашей выбранной страны строим график зараженных в процентах от населения:

```python
pop = countries[(countries['Country_Region']=='Russia') & countries['Province_State'].isna()]['Population'].iloc[0]
df['pinfected'] = df['infected']*100 / pop
df['pinfected'].plot(figsize=(10,3))
plt.show()
```


## Вычисляем заразность $R_t$

Для определения заразности заболевания используют **базовое репродуктивное число** $R_0$, которое показывает число людей, которое в среднем заражает каждый больной индивид. Если $R_0$ больше 1, то эпидемия будет распространяться.

$R_0$ является свойством самого заболевания, и не учитывает защитные меры, принимаемые населением. По мере развития эпидемии, мы можем оценивать такое число $R_t$ в каждый момент времени $t$. Было показано, что можно приблизительно оценить это число, взяв окно в 8 дней, и посчитав $$R_t=\frac{I_{t-7}+I_{t-6}+I_{t-5}+I_{t-4}}{I_{t-3}+I_{t-2}+I_{t-1}+I_t}$$
где $I_t$ - число зараженных в день $t$.

Посчитаем $R_t$ для наших данных о COVID-19. Для этого возьмём **скользящее окно** в 8 дней, и посчитаем на данных из столбца `ninfected` указанное выше отношение:

```python
df['Rt'] = df['ninfected'].rolling(8).apply(lambda x: x[4:].sum()/x[:4].sum())
df['Rt'].plot()
plt.show()
```

В графике видны выколотые интервалы, которые вызваны значениями типа `NaN` или `inf`, возникающими из-за отсутствующих значений в исходном датасете (например, в начале датасета, когда нет возможности посчитать среднее в скользящем окне, т.к. нет начальных значений), или из-за деления на 0. Чтобы сделать график симпатичнее, мы заменим их с использованием функций `replace` и `fillna`.

Посмотрим на начало пандемии поближе. Мы также построим горизонтальную линию в районе 1.

```python
ax = df[(df.index>"2020-03-15") & (df.index<"2020-07-01")]['Rt'].replace(np.inf,np.nan).fillna(method='pad').plot(figsize=(10,3))
ax.set_ylim([0,6])
ax.axhline(1,linestyle='--',color='red')
plt.show()
```

Ещё один интересный индикатор - это **производная**, или **ежедневная разница** в новых случаях. Она позволяет смотреть, в каких случаях пандемия ускоряется или замедляется. 

```python
df['ninfected'].diff().plot()
plt.show()
```

В данных видно много флуктуаций, связанных с особенностями сбора данных. Например, данные в выходные дни существенно отличаются от рабочих дней. Поэтому имеет смысл сгладить данные усреднением со скользящим окном в 7 дней, чтобы выделить тренд:

```python
ax=df['ninfected'].diff().rolling(7).mean().plot()
ax.axhline(0,linestyle='-.',color='red')
plt.show()
```
